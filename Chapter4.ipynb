```python
#Linear regressions: find parameters which minimize the MEAN SQUARE ERROR (equivalently, ROOT MEAN SQUARE ERROR)
#for our given dataset

#There is a closed form solution given by the NORMAL EQUATION

import numpy as np
import matplotlib.pyplot as plt

X = 2*np.random.rand(100,1)
y = 4 + 3*X + np.random.rand(100,1)
```

```python
plt.scatter(X[:],y[:])
```

```python
X_b = np.c_[np.ones((100,1)), X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new,y_predict,'r-')
plt.plot(X,y,'.b')
plt.axis([0,2,0,15])
plt.show()
```
